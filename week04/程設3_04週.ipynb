```python
///week04-1.cpp  模擬題，照著考題寫程式
///LeetCode 挑戰題 3100. Water Bottles II
class Solution {
public:
    int maxBottlesDrunk(int numBottles, int numExchange) {
        int ans = numBottles;///一開始就喝了這麼多瓶
        while(numBottles>=numExchange){///如果瓶子數>=兌換量
            numBottles=numBottles-numExchange+1;///兌換一瓶飲料
            ans++;///多喝一瓶了
            numExchange++;///兌換的標"多一瓶"
        }
    return ans;
    }
};
```

```python
///week04-2.cpp
///LeetCode 學習計畫第10題 896. Monotonic Array
///陣列很單調，只有增加or減少，不可以又增又減
class Solution {
public:
    bool isMonotonic(vector<int>& nums) {
        int up=0,down=0;///up:有增加 down:有減少
        for(int i=1;i<nums.size();i++){///兩兩比較
            if(nums[i-1]<nums[i])up=1;///增加
            if(nums[i-1]>nums[i])down=1;///減少
        }
        if(up==1&&down==1)return false;///不可以又增又減
        return true;///沒有失敗，就成功
    }
};
```
